#### Exercise 1  

Write a program to read through a file and print the contents of the file (line by line) all in upper case.

from [PEW] E8-1

```python
with open('example.txt', 'w') as f:
    f.write('Head Over Heels \n')
    f.write("Something happens and I'm head over heels \n")
    f.write("I never find out 'til I'm head over heels \n")
    f.write("Something happens and I'm head over heels \n")
    f.write("Ah, don't take my heart, don't break my heart \n")
    f.write("Don't, don't, don't throw it away \n")
    f.write("Throw it away, throw it away \n")
```

```python
filename = input("Enter the filename: ")

try:
    with open(filename, 'r') as file_handle:
        for line in file_handle:
            print(line.upper(), end='')
except FileNotFoundError:
    print(filename, "does not exist.")
```

#### Exercise 2  

Write a program to read through a file and output the number of lines in a file.

adapted from [PEW]

```python
filename = input("Enter the filename: ")

try:
    with open(filename, 'r') as file_handle:
        line_count = 0
        for line in file_handle:
            line_count += 1
        print("Number of lines in", filename, ":", line_count)
except FileNotFoundError:
    print(filename, "does not exist.")
```

## CHALLENGE

from [PEW] E8-2

Write a program to prompt for a file name, and then read through the file and look for lines of the form:

```python
X-DSPAM-Confidence: 0.8475
```

When you encounter a line that starts with "X-DSPAM-Confidence:" pull apart the line to extract the floating-point number on the line. Count these lines and then compute the total of the spam confidence values from these lines. When you reach the end of the file, print out the average spam confidence.

```python
Enter the file name: mbox.txt
Average spam confidence: 0.894128046745

Enter the file name: mbox-short.txt
Average spam confidence: 0.750718518519
```

```python
mbox_content = """X-DSPAM-Confidence: 0.8475
X-DSPAM-Confidence: 0.7546
X-DSPAM-Confidence: 0.9587
X-DSPAM-Confidence: 0.7322
X-DSPAM-Confidence: 0.9682
X-DSPAM-Confidence: 0.6745"""

mbox_short_content = """X-DSPAM-Confidence: 0.8475
X-DSPAM-Confidence: 0.7546
X-DSPAM-Confidence: 0.9587"""

with open('mbox.txt', 'w') as mbox_file:
    mbox_file.write(mbox_content)

with open('mbox-short.txt', 'w') as mbox_short_file:
    mbox_short_file.write(mbox_short_content)
```

```python
filename = input("Enter the file name: ")

try:
    with open(filename, 'r') as file_handle:
        total_confidence = 0.0
        count = 0

        for line in file_handle:
            if line.startswith('X-DSPAM-Confidence:'):
                confidence_value = float(line.split(':')[1])
                total_confidence += confidence_value
                count += 1

        if count > 0:
            average_confidence = total_confidence / count
            print("Average spam confidence:", average_confidence)
        else:
            print("No lines starting with 'X-DSPAM-Confidence:' found in the file.")
except FileNotFoundError:
    print("File not found:", filename)
```
