A. Create a four-dimensional array of random integers in the interval [0, 12]. Print the result.

```python
# Creates a four-dimensional array of random integers in the interval [0, 12]
random_array = np.random.randint(0, 15, size=(3, 4, 5, 6))

# Prints the resulting array
print("Four-dimensional array of random integers:")
print(random_array)
```

B. Get the maximum number for each dimensions and display it.

```python
# Get the maximum number for each dimension
max_per_dimension = random_array.max(axis=(0, 1, 2, 3))

# Displays the maximum number for each dimension
print("Maximum number for each dimension:")
print(max_per_dimension)
```

C. Multipy each array with 4. What is its result? * Write a brief explanation on the multiplication process.

```python
# Multiplies each array element by 4. Each element in the array random_array is multiplied by 4, resulting in a new array result where each element is four times its original value.
result = random_array * 4

# Displays the result
print("Result: ")
print(result)
```
