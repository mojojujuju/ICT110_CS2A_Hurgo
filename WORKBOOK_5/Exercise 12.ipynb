A. Create an ndarry object (A) with any values from 1 to 100 in a nested Python List. You can choose your range from the given values (1-100).

```python
import numpy as np   # Imports the numpy module and call it as np

A = np.array([[i for i in range(1, 101)]])  # I used list comprehension to generate values from 1 to 100.
```

1. Using a property of the ndarray object

```python
A_shape = A.shape  # Get the shape of the array
print("Shape of A:", A_shape)
```

2. Using a method of the ndarray object

```python
A_max = A.max()  # Get the maximum value in the array
print("Maximum value in A:", A_max)
```

3. Using a numpy function and passing an ndarray

```python
A_max_np = np.max(A)  # I used np.max() function to find the maximum value.
print("Maximum value in A (using np.max()):", A_max_np)
```

B. Create a 5x5 integer-array filled with ones.

```python
ones_array = np.ones((5, 5), dtype=int)   # Creates a 5x5 integer-array filled with ones

# Checks the shape, size, and number of dimensions
ones_shape = ones_array.shape
ones_size = ones_array.size
ones_ndim = ones_array.ndim

print("Shape of the array:", ones_shape)
print("Size of the array:", ones_size)
print("Number of dimensions of the array:", ones_ndim)
```

C. Create an array of values from -5 to + 5. Multiply each value with 2. Print the resulting array.

```python
array_values = np.arange(-5, 6)   # Creates an array of values from -5 to +5

# Multiply each value with 2
result_array = array_values * 2

print("Result: ", result_array)
```
